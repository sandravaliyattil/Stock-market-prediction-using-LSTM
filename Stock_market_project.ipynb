## Time series plot

```python
plt.figure(figsize=(10, 6))
plt.plot(ts.index, ts.values, label='Original Data')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

From this time series plot, it is clear that the data is not stationary. Also seasonality may occur. We can varify the stationarity condition by ADF Test.

```python
from statsmodels.tsa.stattools import adfuller
close_data = ts['Close']

# Perform the ADF test
result = adfuller(close_data)

print("ADF Statistic: %f" % result[0])
print("p-value: %f" % result[1])
print("Critical Values:")
for key, value in result[4].items():
    print("\t%s: %.3f" % (key, value))
```

```python
if result[1] <= 0.05:
    print("The time series is stationary (reject the null hypothesis)")
else:
    print("The time series is non-stationary (fail to reject the null hypothesis)")
```

```python
# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(close_data, bins=20, edgecolor='k')

plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.title('Histogram of Close Prices')

plt.show()
```

By this histogram, we can conclude that data is not normally distributed.

```python
# Perform seasonal decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts, model='additive')
```

```python
trend = decomposition.trend
seasonal = decomposition.seasonal
residuals = decomposition.resid
```

```python
 #Create subplots for each component
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(10, 8))

# Plot the original time series
axes[0].plot(ts, label='Original')
axes[0].set_ylabel('Original')

# Plot the trend component
axes[1].plot(trend, label='Trend')
axes[1].set_ylabel('Trend')

# Plot the seasonal component
axes[2].plot(seasonal, label='Seasonality')
axes[2].set_ylabel('Seasonality')

# Plot the residuals component
axes[3].plot(residuals, label='Residuals')
axes[3].set_ylabel('Residuals')

# Add titles and legends
axes[0].set_title('Time Series Decomposition')
plt.tight_layout()
plt.show()
```

Trend is increasing. And residuals are stationary, with constant variance and mean zero.

```python
# train/test split
ts_train = ts.iloc[:int(ts.size * .8)]
ts_test = ts.iloc[int(ts.size * .8):]
```

```python
# Plot the data and identify any unusual observations
plt.figure(figsize=(10, 6))
plt.plot(ts_train.index, ts_train.values, 'blue', label='Train Data')
plt.plot(ts_test.index, ts_test.values, 'green', label='Test Data')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

Upto 2022, data is increasing. But from 2022 to 2023 there is a downward trend happening. Again after 2023 data in increasing steadly.

## **Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
# Normalize the data
scaler = MinMaxScaler()
scaler.fit(ts_train.values) # fit on train
scaled_ts_train_values = scaler.transform(ts_train.values) # transform on train
scaled_ts_test_values = scaler.transform(ts_test.values) # transform on test
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

**TimeseriesGenerator** is a class provided by the Keras library, specifically in the **keras.preprocessing.sequence module**. It is designed to generate batches of temporal data for training recurrent neural networks (RNNs), such as LSTM (Long Short-Term Memory) networks. This class is particularly useful for working with time series data in the context of sequence prediction tasks.

The TimeseriesGenerator is used to generate batches of temporal data sequences for training neural networks. It helps in converting time series data into a format suitable for supervised learning.

```python
n_input = 24
n_features = 1
generator = TimeseriesGenerator(scaled_ts_train_values, scaled_ts_train_values, length=n_input, batch_size=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# define model
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

The LSTM layer is designed to handle sequential or time series data. It has 100 units (output shape (None, 100)), which means it outputs a sequence of 100 values for each input sequence.

The number of parameters in the LSTM layer is 40,800, which includes weights and biases. This number is determined by the formula: (4 * units * (units + input_dim + 1)).

The term (4 * units) accounts for the four sets of weights (input, output, forget, and cell state) associated with each LSTM unit.

The LSTM layer processes sequential data and captures patterns across the sequence.

The Dense layer has 101 parameters, including weights and biases.

The total number of trainable parameters in the model is 40,901.

```python
# fit model
model.fit(generator, epochs=50, verbose=0)
```

**generator:**

This parameter is typically a data generator that generates batches of time series data and an LSTM model.

**epochs=50:**

The epochs parameter defines the number of times the entire training dataset will be passed through the neural network. In this case, the model will be trained for 50 epochs.

**verbose=0:**

The verbose parameter controls the verbosity of the training process. A value of 0 means silent mode, where no output is printed during training.

```python
# plot loss
loss_pre_epoch = model.history.history['loss']

plt.plot(range(len(loss_pre_epoch)), loss_pre_epoch)
```

The loss curve converges to a low value and stabilizes, it suggests that the model has learned the patterns in the training data well

```python
test_predictions = []
first_eval_batch = scaled_ts_train_values[-n_input:]

current_batch = first_eval_batch.reshape((1, n_input, n_features))

for i in range(len(ts_test)):

    current_pred = model.predict(current_batch, verbose=0)[0]

    test_predictions.append(current_pred)

    current_batch = np.append(current_batch[:, 1:, :], [[current_pred]], axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
# true_predictions
```

```python
# Plot the data and identify any unusual observations
plt.figure(figsize=(10, 6))
plt.plot(ts_train.index, ts_train.values, 'blue', label='Train Data')
plt.plot(ts_test.index, ts_test.values, 'green', label='Test Data')
plt.plot(ts_test.index, true_predictions, 'orange', label='Forecasted Data')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
# Plot the original data and forecasted values
plt.figure(figsize=(10, 6))
plt.plot(ts_test.index, ts_test.values, label='Original Data')
plt.plot(ts_test.index, true_predictions, label='Forecasted Data')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('ARIMA Forecast')
plt.show()
```
